```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import mpl_toolkits.mplot3d.art3d as art3d
%matplotlib inline
```

```python
stars = pd.read_csv("Estrellas.csv")
```

```python
stars.tail()
```

```python
radio = stars[["Radio (Radios solares)"]].values
temp = stars[["Temperatura (K)"]].values
lum = stars[["Luminosidad en unidades solares"]].values
```

```python
fig = plt.figure(figsize=(4,5))
ax = Axes3D(fig)
z= 4*np.pi*0.000000056704*radio**2*temp**4
ax.plot_surface(radio, temp, z, cmap = "cool")
ax.set_xlabel("Radio (Radios solares)")
ax.set_ylabel("Temperatura (K)")
ax.set_zlabel("Luminosidad (Lum. sol)")
```
